## <font color="maroon"><h4 align="center">Handling Missing Data - fillna, interpolate, dropna: Exercise</font>

## <font color="blue">About the data</font>
- The file **fruits_data.csv** lists the monthly prices of Apple, Banana, Grapes, Mango, and Water Melons from January to October 2022.
- Your task is to fill in the missing values according to the given instructions.

### <font color="purple">**Question1**</font>
- Read the csv file 'fruits_data.csv' into the dataframe 'fruitsDF'.
- show the number of rows and columns in it.
- list all the columns present in the dataframe.
- show the dataframe

```python
# Reading and loading the csv file into 'fruitsDF' dataframe.
fruitsDF = pd.read_csv('Files\\HandlingData_fillna_interpolate_dropna\\fruits_data.csv')
print("The file is loaded and fruitsDF DataFrame is created\n")

# Displaying the ROWS and COLUMNS
print(f'ROWS and COLUMNS of the DataFrame : {fruitsDF.shape}\n')

# Displaying the COLUMNS present in the DataFrame
print(f'COLUMNS present in the DataFrame : {fruitsDF.columns}\n')

# Displaying the DataFrame
fruitsDF
```

### <font color="purple">**Question2**</font>
- Fill all the null values in the dataframe with '-1' and store the result to variable 'new_fruitsDF'.

```python
# Replacing all the null values with -1 and storing it in 'new_fruitsDF' variable
new_fruitsDF = fruitsDF.fillna(-1)

#Showing the updated DataFrame
new_fruitsDF
```

### <font color="purple">**Question3**</font>
Fill the null values in respective columns like below:

    - apple(1kg): Mean value of that column
    - banana(1 dozen): Mean value of that column
    - grapes(1kg): Median value of that column
    - mango(1kg): Median value of that column
    - Water Melons(1): Fill null values with "Not Available".
    
Finally store the output to 'new_fruitsDF' variable

```python
# Filling the null values as mentioned above
new_fruitsDF = fruitsDF.fillna({
    'apple(1kg)' : fruitsDF['apple(1kg)'].mean(),
    'banana(1 dozen)' : fruitsDF['banana(1 dozen)'].mean(),
    'grapes(1kg)' : fruitsDF['grapes(1kg)'].median(),
    'mango(1kg)' : fruitsDF['mango(1kg)'].median(),
    'Water Melons(1)' : "Not Available"
})

# Displaying the updated DataFarme
new_fruitsDF
```

### <font color="purple">**Question4**</font>

- Fill the null values with the non-null value before it (Forward fill method)
- Store the result in 'new_fruitsDF'.

```python
# Filling the null values with values before it - Forward Fill
new_fruitsDF = fruitsDF.fillna(method='ffill')

# Displaying updated DataFrame
new_fruitsDF
```

### <font color="purple">**Question5**</font>
- Drop the rows if it does not contain atleast 4 non-null values (use the concept of thresholding).
- store the result in 'new_fruitsDF'

```python
# Dropping the rows which doesn't contain atleast 4 non-null values using threshold parameter
new_fruitsDF = fruitsDF.dropna(thresh=4)

# Displaying the updated DataFrame
new_fruitsDF
```

### <font color="purple">**Question6**</font>


- Finally, remove the row if it contain null value and store the result in 'new_fruitsDF'
- save that dataframe to csv:'new_fruitsDF.csv' and make sure to set index to False

```python
# Removing the rows if it contains null values
new_fruitsDF = fruitsDF.dropna()

# Displaying the updated DataFarme
new_fruitsDF
```

```python
# Saving the DataFrame into new_fruitsDF.csv file
new_fruitsDF.to_csv('Files\\HandlingData_fillna_interpolate_dropna\\new_fruitsDF.csv', index=False)
```

### <font color="purple">**Question3: Group data using custom function:**</font>
Here, the requirement is to create four groups based on **"imdb_rating"**

- 1 <= imdb_rating <= 3.9 ---> 'Poor'
- 4 <= imdb_rating <= 7.9 ---> 'Average'
- 8 <= imdb_rating <= 10 --->  'Good'
- If none of above satisfy, then return 'others'

```python
# creating a user defined function to group the movies based on their imdb rating
def grouper(idx, df, col):
    '''
        This function returns category based on imdb_rating
    '''
    if 1 <= df[col].loc[idx] <= 3.9 :
        return 'Poor'
    elif 4 <= df[col].loc[idx] <= 7.9 :
        return 'Average'
    elif 8 <= df[col].loc[idx] <= 10 :
        return 'Good'
    else : return 'others'


# Creating groups based on 'imdb_rating'
g = moviesDF.groupby(lambda idx : grouper(idx, moviesDF, 'imdb_rating'))

#g.size()

# Iterating through g and print all the groups
for key, data in g:
    print(f'Group by Key : {key}\n')
    print(data)
    print('\n')
```

```python
# Reading and loading the csv file into 'df_new_movies' dataframe.
df_new_movies = pd.read_csv('Files\\concat_merge\\new_movies.csv')
print("The file is loaded and 'df_new_movies' DataFrame is created\n")

#Showing the new movies DataFrame
df_new_movies
```

- Merge the dataframe df_movies with the df_languages on 'language_id'.

### <font color="purple">**Question4**</font>

```python
# Merging df_movies with df_financials on 'movies_id'. LEFT JOIN
df_movies = pd.merge(df_movies,df_financial, on='movie_id',how='left')

#Showing the top 5 rows of df_movies
df_movies.head()
```
