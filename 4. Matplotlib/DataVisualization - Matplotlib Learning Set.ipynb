# Data Visualization


## Matplotlib - Python library used for data visualization

* pyplot is sub-module inside the **matplotlib** used for plotting the data into the graph

```python
from matplotlib import pyplot as plt
# import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

* Plotiing a **LINE** graph by passing two coordinates 

```python
# X points : x1=0 , x2=6
xPoints = np.array([0,6])

# Y points : y1=0 , y2=256
yPoints = np.array([0,256]) 

# Plotting the line graph using 2 coordinates (x1,y1) & (x2,y2)
plt.plot(xPoints,yPoints) # This plot() function develops the diagram using the inputs

# To display the graph built by plot() we use show()
plt.show()
```

* In order to plot the graph without lines, use the string notation parameter **'o'**, which means **'rings'**

```python
# It just marks the points on the graph
plt.plot(xPoints,yPoints,'o')
plt.show()
```

* Plotting the graph for multiple points

```python
# Plotting the graph for multiple points
xPoints = np.array([1, 2, 6, 8]) # x1=1 , x2=2 , x3=6 , x4=8
yPoints = np.array([3, 8, 1, 10]) # y1=3 , y2=8, y3=1, y4=10

plt.plot(xPoints,yPoints)
plt.show()
```

* Plotting the above graph with :
    - Marker : **Diamond shape**
    - Line : **Dashed-Dotted line**
    - Line Color : **Blue**
    - Marker Size : **25**
    - Maker Edge Color : **Black**
    - Marker Face Color : **Cyan**

```python
# Marker, Line, Line color (fmt) : 'D-.b', where d -> Small diamond, -. -> Dash_sotted Line, b -> Blue Color
# Or Marker : marker = 'D'
# Line style : ls = '-.' or 'dashdot'
# Line color : lc = 'b'
# Marker size : ms = 25
# Marker Edge Color : mec = 'k',  k -> Black color
# Marker Face Color : mfc = 'c',  c -> Cyan color

plt.plot(xPoints,yPoints,'d-.b', ms = 25, mec = 'k', mfc = 'c')
plt.show()
```

* For plotting multiple lines : 
    - Method 1 : Using multiple **plt.plot()**
    - Method 2 : Passing the multiple values in the same **plt.plot()** function

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

# Method 1
plt.plot(x1,y1)
plt.plot(x2,y2)

plt.show()

# Method 2 
plt.plot(x1,y1,x2,y2)

plt.show()
```

* Creating Labels for Plot
    - **xlabel()** : for labeling x-axis
    - **ylabel()** : for labeling y-axis

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x,y)

# Labeling the axis

# X-axis
plt.xlabel("Average Pulse")

# Y-axis
plt.ylabel("Calories Burnage")

plt.show()
```

* Creating Title for the plot
    - **title()** : used to create the title for the plot

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x,y)

# Labels
plt.xlabel("Average Pulse")
plt.ylabel("Calories Burnage")

# Title of the plot
plt.title("Sports Watch Data")

plt.show()
```

* Setting the font properties :
    - **fontdict** parameter is used to set the properties of the fonts

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

# Declaring the font properties for LABEL
fontLabels = {
    'family' : 'serif',
    'color' : 'darkblue',
    'size' : 15
}

# Declaring the font properties for TITLE
fontTitle = {
    'family' : 'serif',
    'color' : 'green',
    'size' : 25
}


# Creting Labels and Title
plt.xlabel("Average Pulse", fontdict=fontLabels)
plt.ylabel("Calories Burnage", fontdict=fontLabels)

plt.title("Sports Watch Data", fontdict=fontTitle)

# Plotting the diagram and showing it
plt.plot(x,y)
plt.show()
```

* Setting the location of TITLE : 
    - **loc** parameter is used to set the location
    - Legal values : **left**, **right**, **center** [Default]

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x,y)

# Positioning the title at the LEFt side of the plot
plt.title("Sports Watch Data", loc='left')

plt.show()
```

* **Grid Lines**
    - **grid()** : method used to plot the grid lines on the graph
    - **which** : to apply changes on {'minor', 'major', 'both}
    - **axis** : which axis the grid lines to come {'x', 'y', 'both'}
    - Can set the **line properties** just like the **graph line**

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

# Labels
plt.xlabel("Average Pulse")
plt.ylabel("Calories Burnage")

# Title
plt.title("Sports Watch Data")

# Plotting the GRID lines on both axis with linestyle : dashed lines( -- ), linewidth : 0.3, color : red
plt.grid(axis='both', linestyle = '--', linewidth = 0.3, color='red')

# Developing and displaying the graph
plt.plot(x,y)
plt.show()
```

* Multiple plots
    - **subplot()** : used to plot multiple plots
    - This function takes 3 arguments, where 1st and 2nd represents the ROW and COLUMN respectively
    - 3rd argument represents the INDEX of the current plot.
    - **suptitle()** : used to set the TITLE for the entire diagram
    - **title()** : used to set the TITLE for individual plot
    

```python
# Plotting 4 graphs with MAIN TITLE : "MY SHOP"
# 1st plot : "SALES", 2nd plot : "INCOME"
# 3rd plot : "REVENUE", 4th plot : "PROFIT"

# plot 1 : SALES
x1 = np.array([0, 1, 2, 3])
y1 = np.array([5, 11, 3, 17])
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title("SALES")

# plot 2 : INCOME
x2 = np.array([0, 1, 2, 3])
y2 = np.array([3, 8, 1, 10])
plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title("INCOME")

# plot 3 : REVENUE
x3 = np.array([0, 1, 2, 3])
y3 = np.array([6, 2, 7, 11])
plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title("REVENUE")

# plot 4 : PROFIT
x4 = np.array([0, 1, 2, 3])
y4 = np.array([10, 20, 30, 40])
plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title("PROFIT")

# MAIN TITLE : MY SHOP
plt.suptitle("MY SHOP")

plt.show()
```

### Scatter Plot

* **scatter()** function is used or pass **kind='scatter'** inside **plot()**
* Leaves dot for each observation
* Requires two arrays of same length as x-axis and y-axis
* For changing color use **color** or **c** inside the function. You can just pass the specific color you want or can pass the colormap with different values.
* There are so many built-in colormaps. Specify them using **cmp** inside the function
* Use **plt.colorbar()** to display the colorbar for representation of the values throigh colors
* Use **alpha** to specify the transparency of the dots inside the fucntion
* Use **s** or **size** inside the function to specify the sizes of the dots. Can pass the single value or an array of values of same length as the axis arrays

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))

# Defining  different color values for the colorbar
colors = np.random.randint(100, size=(100))

#Defining the different sizes for the dots
sizes = 10 * np.random.randint(100, size=(100))

# Plotting the SCATTER graph with color map : nipy_spectral, transparency : 0.5
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

# To display the colorbar with the plot
plt.colorbar()

plt.show()
```

### Bar Graphs

* **bar()** is used to get the **verticle** bars, which takes **width** parameter 
* **barh()** is used to get the **horizontal** bars, which takes **height** parameter
* Takes 2 arguments which represents Categories and their values respectively.
* **color** parameter takes the color name or HEXADECIMAL value for the bar color.


```python
# Categories
x = np.array(["A", "B", "C", "D"])

# Values
y = np.array([3, 8, 1, 10])

# Plotting the vertical bar graph with width : 0.3, color : hotpink
plt.bar(x, y, width=0.3, color='hotpink')
plt.show()

# Plotting the horizontal bar graph with height : 0.4, color : green
plt.barh(x, y, height=0.4, color='lightgreen')
plt.show()
```

### Histogram


* **hist()** or pass **kind='hist'** inside **plot()** to create histograms
* Gives the frequency distribution data
* Takes only one argument


```python
x = np.random.normal(170,10,250)

# Plotting the Histogram
plt.hist(x,rwidth=0.8) # Reducing the width to make each bar visible
plt.show()
```

### Pie Chart

* **pie()** is used to get the pie charts
* Pass an array of names for each wedge using **label** parameter inside the function
* To strat the pie chart from different angle, use **startangle** parameter inside the function
* To make one of the wedge standout from others, then use **explode** parameter inside the function and pass an array of values for each wedge.
* To add shadow to the pie chart, use **shadow** parameter inside the functiion and pass **True** to the parameter
* To change the colors of the wedges, use the **colors** parameter inside the function and pass an array of colors for each wedge.
* To dsiplay an explanation for each wedge, the use **plt.legend()** function.
* To give a title to the legend, use **title** parameter inside the **legend()** function and the value for it

```python
x = np.array([35, 25, 25, 15])

# Labels for each wedges
myLabels = ["Apples", "Bananas", "Cherries", "Dates"]

# Colors for each wedges
myColors = ['r','g','b','c']

# To make the Bananas wedge to standout from others
myExplode = [0,0.3,0,0]

# Plotting the pie chart with shadows ON
plt.pie(x, labels=myLabels, colors=myColors, explode=myExplode, shadow=True)

# Adding the legend at upper right corner with title : FRUITS
plt.legend(loc='upper right',title='FRUITS')

plt.show()
```
